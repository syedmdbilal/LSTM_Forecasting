```python
# Splitting the data into x and y ( input and output )

X=df.drop(columns='Domestic Market (Contract) Blow Molding, Low', axis=1)
y=df.loc[:,'Domestic Market (Contract) Blow Molding, Low']
```

```python
### Train test split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=24)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score, median_absolute_error
```

```python
MLP = MLPRegressor()
MLP.fit(X_train,y_train)
predictMLP = MLP.predict(X_test)

LR = LinearRegression()
LR.fit(X_train,y_train)
predictLR = LR.predict(X_test)

RFR = RandomForestRegressor()
RFR.fit(X_train,y_train)
predictRFR = RFR.predict(X_test)

LIR = LinearRegression()
LIR.fit(X_train,y_train)
predictLIR = LIR.predict(X_test)

LA = Lasso()
LA.fit(X_train,y_train)
predictLA = LA.predict(X_test)

KNR = KNeighborsRegressor()
KNR.fit(X_train,y_train)
predictKNR = KNR.predict(X_test)

SV = SVR()
SV.fit(X_train,y_train)
predictSV = SV.predict(X_test)
```

```python
MAE = (mean_absolute_error(y_test,predictMLP))
print("MEAN ABSOLUTE ERROR OF MLP:", MAE)
print('')

MAE = (mean_absolute_error(y_test,predictLR))
print("MEAN ABSOLUTE ERROR OF LR:", MAE)
print('')

MAE = (mean_absolute_error(y_test,predictRFR))
print("MEAN ABSOLUTE ERROR OF RFR:", MAE)
print('')

MAE = (mean_absolute_error(y_test,predictLIR))
print("MEAN ABSOLUTE ERROR OF LIR:", MAE)
print('')

MAE = (mean_absolute_error(y_test,predictLA))
print("MEAN ABSOLUTE ERROR OF LA:", MAE)
print('')

MAE = (mean_absolute_error(y_test,predictKNR))
print("MEAN ABSOLUTE ERROR OF KNR:", MAE)
print('')

MAE = (mean_absolute_error(y_test,predictSV))
print("MEAN ABSOLUTE ERROR OF SVR:", MAE)
```

```python
MSE = (mean_squared_error(y_test,predictMLP))
print("MEAN SQUARED ERROR OF MLP:", MSE)
print('')

MSE = (mean_squared_error(y_test,predictLR))
print("MEAN SQUARED ERROR OF LR:", MSE)
print('')

MSE = (mean_squared_error(y_test,predictRFR))
print("MEAN SQUARED ERROR OF RFR:", MSE)
print('')

MSE = (mean_squared_error(y_test,predictLIR))
print("MEAN SQUARED ERROR OF LIR:", MSE)
print('')

MSE = (mean_squared_error(y_test,predictLA))
print("MEAN SQUARED ERROR OF LA:", MSE)
print('')

MSE = (mean_squared_error(y_test,predictKNR))
print("MEAN SQUARED ERROR OF KNR:", MSE)
print('')

MSE = (mean_squared_error(y_test,predictSV))
print("MEAN SQUARED ERROR OF SVR:", MSE)
```
